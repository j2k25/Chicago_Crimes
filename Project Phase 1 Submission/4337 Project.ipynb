```python
%%time

# 0   ID                    int64  
#  1   Case Number           object 
#  2   Date                  object 
#  3   Block                 object 
#  4   IUCR                  object 
#  5   Primary Type          object 
#  6   Description           object 
#  7   Location Description  object 
#  8   Arrest                bool   
#  9   Domestic              bool   
#  10  Beat                  int64  
#  11  District              float64
#  12  Ward                  float64
#  13  Community Area        float64
#  14  FBI Code              object 
#  15  X Coordinate          float64
#  16  Y Coordinate          float64
#  17  Year                  int64  
#  18  Updated On            object 
#  19  Latitude              float64
#  20  Longitude             float64
#  21  Location              object 

# 18 for later use
chi = Chi01P.iloc[:,[5,7,8,9,10,11,12,13,17]]

chi.info()
display(chi.nunique())
print('Primary Type ')
display(chi['Primary Type'].unique())
print('Beat')
display(chi['Beat'].unique())
print('District ')
display(chi['District'].unique())
print('Ward ')
display(chi['Ward'].unique())
print('Community Area ')
display(chi['Community Area'].unique())
print('Year')
display(chi['Year'].unique())
display(chi.head())
```

```python
%%time

chi.where(chi.iloc[:,[3]] == 0, other=np.NaN, inplace=True, axis=None, level=None, errors='ignore', try_cast=False)
                                                                                          # raise

# display(chi.iloc[:,[3]])

chi.info()
chi
```

```python
%%time
chi.info()

#     Primary Type           
#     Location Description   
#     Arrest                
#     Domestic              
#     Beat                    
#     District              
#     Ward                  
#     Community Area        
#     Year                     

chi = pd.DataFrame(chi.dropna().to_numpy(), columns=['Primary Type','Location Description','Arrest',
                                                     'Domestic','Beat','District','Ward',
                                                     'Community Area','Year'])                      

chi.pop('Domestic')

chi = chi.convert_dtypes()
chi.info()
chi
```

```python
%%time

chi.sort_values("Year", axis = 0, ascending = True, inplace = True, na_position ='last', ignore_index=True)
chi
```

```python
%%time

chi['Location Description'] = chi['Location Description'].str.split().str[0]
chi['Location Description'] = chi['Location Description'].str.split('/').str[0]

print(chi['Location Description'].unique())
chi.info()
print(chi.nunique())
chi
```

```python
%%time

# Import label encoder
from sklearn import preprocessing 
  
# label_encoder object knows how to understand word labels. 
le = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
chi['Location Description']= le.fit_transform(chi['Location Description']) 
                                        #   inverse_transform(y)
print(chi['Location Description'].unique())
# display(le.get_params(deep=True))
chi.info()
chi
```
